```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

```python
# parámetros
n_samples = 1000
n_features = 10

# dataset sintético
rng = np.random.default_rng(42)
X = pd.DataFrame(rng.normal(size=(n_samples, n_features)),
                 columns=[f"x{i}" for i in range(n_features)])
y = pd.DataFrame(rng.uniform(size=(n_samples, 1)), columns=["target"])

# split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# 1. Build and compile model


```python
model = Sequential() # creating the sequential model

# now adding the layers to the model
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns))) # dense layer with 32 neurons and relu activation 
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossetropy', optimizer='sgd', metrics='accuracy')  # how to train the model
```

# 2. Fit, predict and evaluate

```python
model.fit(X_train, y_train, epochs=200, batch_size=32) # training the model
```

```python
y_hat = model.predict(X_test)
y_hat = [0 if val <0.5 else 1 for val in y_hat]
```

```python
y_hat
```

```python
accuracy_score(y_test, y_hat)
```

# 3. Save and reload

```python
model.save('test_model')
```

```python
del model # deleting the existing model from memory
```

```python
model = load_model('test_model') # loading the saved model
```
